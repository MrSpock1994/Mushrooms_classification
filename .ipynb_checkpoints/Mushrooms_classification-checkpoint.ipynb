```python
import pandas as pd
```

```python
mush = pd.read_csv("mushrooms.csv")
```

```python
mush.info()
```

```python
mush["class"].unique()
```

```python
mush
```

```python
mask1 = mush["class"] == "p"
mush.loc[mask1, "class"] = 1
mask2 = mush["class"] == "e"
mush.loc[mask2, "class"] = 0
mush
```

```python
mush = mush.astype("category")
```

```python
mush["class"] = mush["class"].astype("int")
```

```python
mush.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
mush['cap-shape'] = labelencoder.fit_transform(mush['cap-shape'])
mush
```

```python
colums = mush.columns
```

```python
colums
for col in colums:
    mush[col]= labelencoder.fit_transform(mush[col])
mush
```

```python
mush = mush.astype("int")
```

```python
x_features = mush.drop("class", axis = 1)
```

```python
y_target = mush["class"].astype("int")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#we are going to split the data into 70% training and 30% testing
x_train, x_test, y_train, y_test = train_test_split(x_features, y_target, test_size=0.3)
```

```python
model = LogisticRegression(max_iter=100000)
model.fit(x_train, y_train)
```

```python
print(x_train.shape, x_test.shape)
```

```python
predictions = model.predict(x_test)
accuracy_score(y_test, predictions)
```
